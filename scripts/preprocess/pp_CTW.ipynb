```python
import os, sys, glob
from os.path import join as ospj

data_root = "/data/zhaoym/UDiffText/CTW"
image_root = ospj(data_root, "image")
anno_root = ospj(data_root, "annotation")
os.makedirs(image_root, exist_ok=True)
os.makedirs(anno_root, exist_ok=True)

tar_files = sorted(glob.glob(ospj(data_root, "*.tar")))
anno_tar_file = ospj(data_root, "ctw-annotations.tar.gz")
tar_files
```

```python
for tar_file in tar_files:
    os.system(f"tar -xf {tar_file} -C {image_root}")

os.system(f"tar -xzf {anno_tar_file} -C {anno_root}")
```

```python
import json
train_anno_path = ospj(anno_root, "train.jsonl")
lines = []
annos = []
with open(train_anno_path) as fp:
    lines = fp.readlines()
for line in lines:
    anno = json.loads(line)
    annos.append(anno)
len(annos)
```
