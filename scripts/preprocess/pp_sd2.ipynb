```python
import torch

name = "512-inpainting-ema"

ckpt = torch.load(f"./checkpoints/pretrained/{name}.ckpt", map_location="cpu")
sd = ckpt["state_dict"]
keys = list(sd.keys())
keys
```

```python
for key in keys:
    if not key.startswith("model.diffusion_model"):
        print(f"Del key: {key}")
        sd.pop(key)
    elif "attn2" in key or "norm2" in key:
        print(f"Del key: {key}")
        sd.pop(key)
```

```python
keys = list(sd.keys())
for key in keys:
    if not "output_block" in key and not "diffusion_model.out" in key:
        ctrl_key = key.replace("diffusion_model", "diffusion_model.ctrl_model")
        print(f"New_key: {ctrl_key}")
        sd[ctrl_key] = sd[key].clone()
```

```python
ckpt["state_dict"] = sd
torch.save(ckpt, f"./checkpoints/pretrained/{name}-ctrl.ckpt")
```
