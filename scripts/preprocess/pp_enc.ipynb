```python
import torch

name = "512-inpainting-ema"

ckpt = torch.load(f"./checkpoints/encoders/EMD/model.ckpt", map_location="cpu")
sd = ckpt["state_dict"]
keys = [key for key in sd.keys() if key.startswith("cond_stage_model")]
keys
```

```python
torch.save(new_sd, f"./checkpoints/encoders/EMD/model.pth")
```
